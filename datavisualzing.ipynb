```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


grouped_data = youtube.groupby(['channel_type', 'Country'])['Youtuber'].count().reset_index()

# Pivot the data for heatmap
pivot_data = grouped_data.pivot('channel_type', 'Country', 'Youtuber')

# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(pivot_data, annot=True, fmt="g", cmap="YlGnBu")
plt.title('YouTubers by Channel Type and Country')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd


# Create a DataFrame from the data
df = pd.DataFrame(youtube1_dict.keys(), columns=["created_year", "Title"])
df["Count"] = youtube1_dict.values()

# Get unique years
unique_years = df["created_year"].unique()

# Calculate the number of subplots needed
num_subplots = len(unique_years)

# Create subplots
fig, axes = plt.subplots(nrows=num_subplots, ncols=1, figsize=(10, 5 * num_subplots))

# Loop through each unique year and create a subplot
for i, year in enumerate(unique_years):
    ax = axes[i]
    data_subset = df[df["created_year"] == year]
    data_subset.plot(kind='bar', x='Title', y='Count', ax=ax, color='dodgerblue')
    ax.set_title(f'Channel Type Counts for Year {year}')
    ax.set_xlabel('Title')
    ax.set_ylabel('Count')
    ax.legend().remove()

# Adjust layout
plt.tight_layout()

# Show the subplots
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd


# Create a DataFrame from the data
df = pd.DataFrame(youtube1_dict.keys(), columns=["created_year", "Title"])
df["Count"] =youtube1_dict.values()

# Choose the year you want to analyze
target_year = 2022.0

# Filter data for the target year
target_data = df[df["created_year"] == target_year]

# Calculate total counts for each title
title_totals = target_data.groupby("Title")["Count"].sum()

# Create a bar plot for total counts
title_totals.plot(kind='bar', color='dodgerblue', figsize=(10, 6))
plt.xlabel('Title')
plt.ylabel('Total Count')
plt.title(f'Total Channel Type Counts for Year {target_year}')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python

# Aggregate data by year and title to calculate total subscribers
grouped_data = youtube.groupby(['created_year', 'Title'])['subscribers'].sum().reset_index()

# Create a line plot for each channel's total subscribers over the years
plt.figure(figsize=(12, 6))
for title in df['Title'].unique():
    channel_data = grouped_data[grouped_data['Title'] == title]
    plt.plot(channel_data['created_year'], channel_data['subscribers'], marker='o', label=title)

plt.xlabel('Year')
plt.ylabel('Total Subscribers')
plt.title('Total Subscribers for Each Channel Over the Years')
plt.legend()
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```

```python

grouped_data = youtube.groupby('category')['subscribers_for_last_30_days'].sum().reset_index()

# Create a bar plot for total subscribers for each category
plt.figure(figsize=(10, 6))
plt.bar(grouped_data['category'], grouped_data['subscribers_for_last_30_days'], color='dodgerblue')
plt.xlabel('Category')
plt.ylabel('Total Subscribers (Last 30 Days)')
plt.title('Total Subscribers for Each Category (Last 30 Days)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Show the plot
plt.show()
```

```python

grouped_data = youtube.groupby('category')['subscribers_for_last_30_days'].sum().reset_index()

# Sort the data by subscribers in descending order
sorted_data = grouped_data.sort_values(by='subscribers_for_last_30_days', ascending=False)

# Define colors for the highest and lowest values
colors = ['dodgerblue' if x == sorted_data.iloc[0]['category'] else 'orange' for x in sorted_data['category']]

# Create a bar plot for total subscribers for each category
plt.figure(figsize=(10, 6))
bars = plt.bar(sorted_data['category'], sorted_data['subscribers_for_last_30_days'], color=colors)
plt.xlabel('Category')
plt.ylabel('Total Subscribers (Last 30 Days)')
plt.title('Total Subscribers for Each Category (Last 30 Days)')
plt.xticks(rotation=45, ha='right')

# Adding legend
legend_labels = ['Highest Subscribers', 'Other Categories']
plt.legend(bars[:2], legend_labels)

plt.tight_layout()

# Show the plot
plt.show()
```

```python

grouped_data = youtube.groupby('category')['subscribers_for_last_30_days'].sum().reset_index()

# Find the maximum and minimum values
max_value = grouped_data['subscribers_for_last_30_days'].max()
min_value = grouped_data['subscribers_for_last_30_days'].min()

# Define colors for the highest and lowest values
colors = ['dodgerblue' if x == max_value else 'orange' if x == min_value else 'gray' for x in grouped_data['subscribers_for_last_30_days']]

# Create a bar plot for total subscribers for each category
plt.figure(figsize=(10, 6))
bars = plt.bar(grouped_data['category'], grouped_data['subscribers_for_last_30_days'], color=colors)
plt.xlabel('Category')
plt.ylabel('Total Subscribers (Last 30 Days)')
plt.title('Total Subscribers for Each Category (Last 30 Days)')
plt.xticks(rotation=45, ha='right')

# Adding legend
legend_labels = ['Highest Subscribers', 'Lowest Subscribers', 'Other Categories']
plt.legend(bars[:3], legend_labels)

plt.tight_layout()

# Show the plot
plt.show()
```
